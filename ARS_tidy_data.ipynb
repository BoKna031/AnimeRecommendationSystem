```python
import os

DATASET_FOLDER = os.sep + 'content'+ os.sep +'drive'+ os.sep +'MyDrive'+ os.sep +'Colab Notebooks'+ os.sep +'dataset'
GENERATED_FOLDER = DATASET_FOLDER + os.sep + 'generated'

ANIME_CSV_PATH = DATASET_FOLDER + os.sep + 'anime.csv'
ANIME_ADDINFO_CSV_PATH = DATASET_FOLDER + os.sep + 'anime_with_synopsis.csv'
#BIG_RATINGS_CSV_PATH = DATASET_FOLDER + os.sep + 'animelist.csv'
SMALL_RATING_CSV_PATH = DATASET_FOLDER + os.sep + 'rating.csv'
```

defining paths to be used for generating and reading csv files

all the files that make up the basic dataset will be in the **dataset** folder, while the generated files will be in the **dataset/generated** folder

```python
import pandas as pd
anime_df = pd.read_csv(ANIME_CSV_PATH)
anime_df.head()
```

```python
anime_df['Genres'] = anime_df['Genres'].str.split(', ')
anime_df.info()
```

```python
anime_df.shape[0]
anime_df.head()
```

```python
unique_genres = anime_df['Genres'].explode().unique()
unique_genres.sort()
print(unique_genres)
```

Inappropriate genres as *Hentai* should be removed.

```python
inappropriate_content = ['Hentai']
mask = anime_df['Genres'].apply(lambda content: any(inap_cont in content for inap_cont in inappropriate_content))
inappropriate_df_index = anime_df[mask].index
mal_ids = anime_df.loc[inappropriate_df_index, 'MAL_ID']
anime_df = anime_df.drop(inappropriate_df_index)
print('removed rows: ', len(mal_ids))
```

```python
unique_genres = anime_df['Genres'].explode().unique()
unique_genres.sort()
print(unique_genres)
```

Now we can try to find data about *Unknown* animes from other available datasets

```python
anime_with_unknown_genres = anime_df.loc[anime_df['Genres'].apply(lambda x: 'Unknown' in x)]
print(anime_with_unknown_genres.shape[0])
```

```python
anime_with_unknown_genres.head()
```

There is a small number of animes that contains *Unknown* as genre, we can manually enter that information. Because there is problem with showing list of all animes (interactive data table shows only first 5 records), I will create new csv file. 

```python
ANIME_WITH_UNKNOWN_GENRES_CSV_PATH = GENERATED_FOLDER + os.sep + 'unknown_genres_anime.csv'
anime_with_unknown_genres.to_csv(ANIME_WITH_UNKNOWN_GENRES_CSV_PATH, index = False)
```
