# Group anime by series

To group anime by series we need to find related animes (same anime but different season)

Calcualtion will be divided in 3 steps:
1.   Comparing name of animes using Cosine Similarity
2.   Comparing genres of animes using Jaccard Similarity
3. Comparing studio of animes using Jaccard Similarity

For each of the step, we will choose a special coefficient that we will add to the equation by which we calculate the overall similarity


```python
import math 
import re
from collections import Counter

def text_to_vector(text):
    words = re.compile(r'\w+').findall(text)
    return Counter(words)

def cosine_similarity(str1, str2):
    vector1 = text_to_vector(str1)
    vector2 = text_to_vector(str2)
    intersection = set(vector1.keys()) & set(vector2.keys())
    numerator = sum([vector1[x] * vector2[x] for x in intersection])

    sum1 = sum([vector1[x]**2 for x in vector1.keys()])
    sum2 = sum([vector2[x]**2 for x in vector2.keys()])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)

    if not denominator:
        return 0.0
    else:
        return float(numerator) / denominator

def jaccard_sim(set1, set2): 
    a = set1
    b = set2
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
a1 = "Attack on Titan: Final Season - The Final Chapters"
a2 = "Attack on Titan"
cosine_similarity(a1, a2)
```

```python
def union(x):
    str1 = str(x['Producers'])
    str2 = str(x['Studios'])
    creators_str = str1 + ', ' + str2
    items = creators_str.split(', ')
    return set(items)

anime_df['Creators'] = anime_df.apply(union, axis=1)
```

```python
Name, Genres, Creators = 1,3,4
```

```python
def similarity_score(anime1, anime2):
  c1, c2, c3 = 8, 1, 1
  name_cos_sim =  cosine_similarity(anime1[Name], anime2[Name])
  creators_jac_sim = jaccard_sim(anime1[Creators], anime2[Creators])
  gen_jac_sim = jaccard_sim(set(anime1[Genres]), set(anime2[Genres]))
  score = c1 * name_cos_sim + c2 * creators_jac_sim + c3 * gen_jac_sim
  score = score / (c1 + c2 + c3)
  return round(score,3)
```

```python
def create_similarity_matrix(df):
  df_lst = df.values.tolist()
  num_anime = df.shape[0]
  prog_inc = num_anime // 10
  similarity_matrix = np.zeros((num_anime, num_anime))
  for i, anime1 in enumerate(df_lst):
      for j in range(i+1, num_anime):
          similarity_matrix[i, j] = similarity_score(anime1, df_lst[j])
      if i % prog_inc == 0:
        print(i // prog_inc * 10, '%')
  return similarity_matrix
```

```python
anime_df = anime_df.reset_index(drop=True)
animes_test = anime_df
display_columns = ['MAL_ID','Name', 'English name', 'Genres', 'Creators']
animes_test = animes_test.drop(columns=[col for col in animes_test.columns if col not in display_columns])
animes_test = animes_test.reindex(columns = display_columns)
```

```python
import numpy as np
similarity_matrix_path = "similarity_matrix_tuple.txt"
anime_df_slice = animes_test
try:
  similarity_matrix = np.loadtxt(similarity_matrix_path, delimiter=',')
except FileNotFoundError:
  print('File not found => create new similarity matrix')
  similarity_matrix = create_similarity_matrix(anime_df_slice)
  np.savetxt(similarity_matrix_path, similarity_matrix, delimiter=",")
```

```python
closedGroups = list(nx.connected_components(graph))

anime_series_df = pd.DataFrame(columns = ['Name','MAL_ID','Group'])
for g_ind, group in enumerate(closedGroups):
  anime_name = []
  anime_id = []
  anime_group = []
  for anime_index in group:
    anime = animes_test.iloc[anime_index]
    anime_name.append(anime['Name'])
    anime_id.append(anime['MAL_ID'])
    anime_group.append(g_ind)
  temp_df = pd.DataFrame({'Name':anime_name, 'MAL_ID':anime_id, 'Group': anime_group})
  anime_series_df = pd.concat([anime_series_df, temp_df], ignore_index = True)
```
